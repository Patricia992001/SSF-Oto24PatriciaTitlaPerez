```python
import matplotlib.pyplot as plt

# Datos
altura_central = [
    1.6599955, 1.66999, 1.6799845, 1.689979, 1.6999735,
    1.709968, 1.7199625, 1.729957, 1.7399515, 1.749946
]
frecuencia1 = [96, 97, 97, 93, 115, 113, 107, 97, 95, 90]
frecuencia2 = [100.47, 100.16, 98.63, 98.78, 100.12, 102.85, 99.31, 99.55, 101.19, 98.94]
frecuencia3 = [99.8931, 100.1169, 100.0106, 99.9729, 100.0566, 99.8783, 100.0912, 99.9862, 99.985, 100.0092]

# Crear la figura y los ejes
plt.figure(figsize=(10, 6))

# Gráfica de dispersión
plt.scatter(altura_central, frecuencia1, label='Frecuencia 1', color='blue', marker='o')
plt.scatter(altura_central, frecuencia2, label='Frecuencia 2', color='red', marker='x')
plt.scatter(altura_central, frecuencia3, label='Frecuencia 3', color='green', marker='^')

# Etiquetas y título
plt.xlabel('Altura Central (bin X)')
plt.ylabel('Frecuencia')
plt.title('Gráfico de Dispersión de Frecuencias')
plt.legend()

# Mostrar la gráfica
plt.grid(True)
plt.show()
```
