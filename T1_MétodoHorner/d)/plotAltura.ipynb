### Datos de altura de hombres mexicanos 


El siguiente codigo muestra una grafica de dispersión. Es ideal para observar la distribución de los datos en relación con la altura central. Cada punto representa una frecuencia en una altura central específica y permite ver si hay algún patrón o tendencia no lineal entre los datos.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

# Cargar datos
datos = np.loadtxt("C:/Users/Hp/Desktop/SSF-Oto24PatriciaTitlaPerez/T1_MétodoHorner/d)/altura5.dat")

x = datos[:, 0]
y1 = datos[:, 1]
y2 = datos[:, 2]
y3 = datos[:, 3]

# Crear una figura
plt.figure(figsize=(10, 6))

# Interpolación
x_new = np.linspace(x.min(), x.max(), 500)  # Nuevos puntos de x para interpolar

# Interpolación para cada conjunto de datos
interp1 = interp1d(x, y1, kind='cubic')
interp2 = interp1d(x, y2, kind='cubic')
interp3 = interp1d(x, y3, kind='cubic')

# Gráfica de dispersión
plt.scatter(x, y1, label='Frecuencia 1', color='blue', marker='o')
plt.scatter(x, y2, label='Frecuencia 2', color='red', marker='x')
plt.scatter(x, y3, label='Frecuencia 3', color='green', marker='^')

# Gráfica de las líneas interpoladas
plt.plot(x_new, interp1(x_new), color='blue', linestyle='--')
plt.plot(x_new, interp2(x_new), color='red', linestyle='--')
plt.plot(x_new, interp3(x_new), color='green', linestyle='--')

# Etiquetas 
plt.xlabel('Altura Central (bin X)')
plt.ylabel('Frecuencia')
plt.title('Grafico de dispersión e interpolación de frecuencias')
plt.suptitle('Datos de alturas de hombres mexicanos')
plt.legend()

# Mostrar la gráfica
plt.grid(True)
plt.show()
```
