Ejemplo 4.1.3: Escribe un programa que trace el movimiento de una masa que oscila al final de un resorte. La fuerza sobre la masa debe estar dada por
$$ F = −mg + kx$$
\
Empleando el método de Euler 

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros del sistema
m = 1.0    # Masa (kg)
k = 10.0   # Constante del resorte (N/m)
g = 9.81   # Aceleración debida a la gravedad (m/s^2)
x0 = 0.1   # Posición inicial (m)
v0 = 0.0   # Velocidad inicial (m/s)
dt = 0.01  # Paso de tiempo (s)
t_max = 10.0  # Tiempo máximo de simulación (s)

# Inicialización de las variables
t = np.arange(0, t_max, dt)
x = np.zeros_like(t)
v = np.zeros_like(t)
a = np.zeros_like(t)

# Condiciones iniciales
x[0] = x0
v[0] = v0

# Simulación del movimiento usando el método de Euler
for i in range(1, len(t)):
    # Calcular la fuerza sobre la masa
    F = -m * g + k * x[i-1]
    
    # Calcular la aceleración
    a[i] = F / m
    
    # Actualizar la velocidad y la posición
    v[i] = v[i-1] + a[i] * dt
    x[i] = x[i-1] + v[i] * dt

# Graficar los resultados
plt.figure(figsize=(12, 9))

plt.subplot(3, 1, 1)
plt.plot(t, x, label='Posición (m)')
plt.title('Movimiento de una masa en un resorte')
plt.xlabel('Tiempo (s)')
plt.ylabel('Posición (m)')
plt.legend()
plt.grid()

plt.subplot(3, 1, 2)
plt.plot(t, v, label='Velocidad (m/s)', color='orange')
plt.xlabel('Tiempo (s)')
plt.ylabel('Velocidad (m/s)')
plt.legend()
plt.grid()

plt.subplot(3, 1, 3)
plt.plot(x, v, label='Diagrama de fase (Velocidad vs Posición)', color='green')
plt.xlabel('Posición (m)')
plt.ylabel('Velocidad (m/s)')
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```

$$F= −mg − kx $$
