### 4) Integrar numéricamente

#### Método del trapecio

### Función 1

$$ \boxed{f(x)=\int \limits_{0}^{5} \frac{e^x} {x}dx}$$

```python
import numpy as np
from scipy.integrate import quad
import matplotlib.pyplot as plt

# Parámetros de la integración
a = 1 # Límite inferior
b = 5  # Límite superior
n = 40  # subintervalos

# Definimos la función
def int(x):
    return np.exp(x) / x

# Método del trapecio
def int_trap(f, a, b, n):
    h = (b - a) / n  # Ancho de los subintervalos
    sum_ = 0.5 * (f(a) + f(b))  

    # Sumamos las áreas de los trapecios
    for i in range(1, n):
        x = a + i * h
        sum_ += f(x)

    return sum_ * h

# Calculamos la integral
result = int_trap(int, a, b, n)

# Calcular la integral exacta usando scipy.integrate.quad
valor_exacto, _ = quad(int, a, b)

print(f"Resultado exacto:{valor_exacto}")
print(f"Resultado con método del trapecio: {result}")


# Graficar la integral exacta
x = np.linspace(a, b + 1, 1000)  
y = int(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, color='blue')
plt.fill_between(x, y, alpha=0.2, color='blue')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

### Función 2

$$ \boxed{f(x)=\int \limits_{0}^{5} \frac{1-e^x} {x}dx}$$

```python
import numpy as np
from scipy.integrate import quad
import matplotlib.pyplot as plt

# Parámetros de la integración
a = 1  # Límite inferior
b = 5  # Límite superior
n = 1000  # subintervalos

# Definimos la función
def int(x):
    return (1-np.exp(x)) / x

# Método del trapecio
def int_trap(f, a, b, n):
    h = (b - a) / n  # Ancho de los subintervalos
    sum_ = 0.5 * (f(a) + f(b))  

    # Sumamos las áreas de los trapecios
    for i in range(1, n):
        x = a + i * h
        sum_ += f(x)

    return sum_ * h

# Calculamos la integral
result = int_trap(int, a, b, n)

# Calcular la integral exacta usando scipy.integrate.quad
valor_exacto, _ = quad(int, a, b)

print(f"Resultado exacto:{valor_exacto}")
print(f"Resultado con método del trapecio: {result}")


# Graficar la integral exacta
x = np.linspace(a, b + 1, 1000)  
y = int(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, color='blue')
plt.fill_between(x, y, alpha=0.2, color='blue')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```
