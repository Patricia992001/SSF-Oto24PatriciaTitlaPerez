$\textbf{A2:}$ Estudie una función para diferentes valores de parámetros. 
Desarrolle un programa que cree una gráfica de la función $f(x)=sin \frac{1}{x+\epsilon}$ para $x$ en el intervalo unitario, donde $ \epsilon > 0 $ es un parámetro de entrada dado. Utilice $n + 1$ nodos en la gráfica.

$\textbf{(a)}$ Pruebe el programa usando $n = 10$ y $\epsilon=1/5$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos la función f(x)
def f(x, epsilon):
    return np.sin(1 / (x + epsilon))

# Parámetros
n = 10  # Número de subintervalos
epsilon = 1/5  # Parámetro epsilon
x_values = np.linspace(0, 1, n+1)  # n+1 nodos en el intervalo [0, 1]

# Evaluamos la función para cada valor de x
y_values = f(x_values, epsilon)

# Graficamos
plt.plot(x_values, y_values, marker='o')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title(f'f(x) = sin(1/(x + epsilon)) con epsilon={epsilon}')
plt.grid(True)
plt.show()
```

$\textbf{(b)}$ Redefina el programa de tal manera que trace la función para dos valores de $n$; Para $n$ y $n+10$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos la función f(x)
def f(x, epsilon):
    return np.sin(1 / (x + epsilon))

# Parámetros
n = 10  # Número de subintervalos
epsilon = 1/5  # Parámetro epsilon

# Valores de x para n y n+10
x_values_n = np.linspace(0, 1, n+1)  # n+1 nodos en el intervalo [0, 1]
x_values_n10 = np.linspace(0, 1, n+11)  # (n+10)+1 nodos en el intervalo [0, 1]

# Evaluamos la función para cada valor de x
y_values_n = f(x_values_n, epsilon)
y_values_n10 = f(x_values_n10, epsilon)

# Grafica
plt.plot(x_values_n, y_values_n, marker='o', label=f'n = {n}')
plt.plot(x_values_n10, y_values_n10, marker='x', label=f'n = {n+10}')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title(f'f(x) para n y n+10 con epsilon={epsilon}')
plt.grid(True)
plt.legend()
plt.show()
```

¿Qué tan grande tienes que elegir $n$ para que la diferencia entre estas dos funciones sea menor que $0.1$? Sugerencia: Cada función da una matriz. Cree un bucle while y utilice la función max de las matrices para recuperar el valor máximo y compararlos.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos la función f(x)
def f(x, epsilon):
    return np.sin(1 / (x + epsilon))

# Parámetros 
epsilon = 1/5  # Valor de epsilon
tolerance = 0.1  # Diferencia máxima permitida

# Inicializamos n
n = 10

# Bucle para encontrar el n adecuado
max_difference = tolerance + 1  # Inicializamos con un valor mayor a la tolerancia

while max_difference > tolerance:
    # Generamos los valores de x para n y n+10
    x_values_n = np.linspace(0, 1, n+1)
    x_values_n10 = np.linspace(0, 1, n+11)
    
    # Evaluamos las funciones
    y_values_n = f(x_values_n, epsilon)
    y_values_n10 = f(x_values_n10, epsilon)
    
    # Interpolamos los valores de y_values_n10 a los mismos puntos de x_values_n para comparar
    y_values_n10_interp = np.interp(x_values_n, x_values_n10, y_values_n10)
    
    # Calculamos la diferencia máxima entre las dos funciones
    max_difference = np.max(np.abs(y_values_n - y_values_n10_interp))
    
    # Incrementamos n si la diferencia aún no es menor a la tolerancia
    if max_difference > tolerance:
        n += 1

# Mostramos el valor de n 
print(f'El valor mínimo de n para que la diferencia máxima sea menor que {tolerance} es: {n}')

# Grafica
plt.plot(x_values_n, y_values_n, marker='o', label=f'n = {n}')
plt.plot(x_values_n10, y_values_n10, marker='x', label=f'n = {n+10}')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title(f'Plot of f(x) = sin(1/(x + epsilon)) with epsilon={epsilon}')
plt.grid(True)
plt.legend()
plt.show()
```

$\textbf{(e)}$ Trate de encontrar una fórmula para determinar qué tan grande debe ser n para un valor dado de ε tal que aumentar n aún más no cambie tanto el gráfico como para que sea visible en la pantalla. Tenga en cuenta que no hay una respuesta exacta a esta pregunta.


la función $f(x)=sin \frac{1}{x+\epsilon}$ tiene un comportamiento oscilatorio, y la frecuencia de esas oscilaciones aumenta a medida que $𝑥$ se acerca a $0$. A medida que $\epsilon$ disminuye, el denominador en se hace mas pequeña, lo que incrementa la frecuencia de las oscilaciones, por lo tanto, necesitamos incremar los nodos $(n)$ en el intervlo $[0,1]$ para que la diferencian entre las curvas no sea visible.
Sea $K$ una constante ajustable que depende de la presición deseada. El valor de $n$ necesario para que la diferencia entre las curvas no sea visible depende aproximadamente de la relación inversa con $\epsilon$ es decir:

$$ n \sim \frac{k}{\epsilon}$$
