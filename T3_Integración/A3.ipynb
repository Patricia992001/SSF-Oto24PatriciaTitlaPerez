### Estudia una función y su derivada.

Concidere la función 
$$f(x)= sin \frac{1}{x+\epsilon}$$
para $x$ comprendido entre $[0,1]$ y la derivada 
$$f´(x)=\frac{-cos {\frac{1}{x+\epsilon} }}{(x+\epsilon)^2}

Desarrolle un programa que cree una gráfica de la derivada de $f = f(x)$ basada en una aproximación de diferencias finitas utilizando n nodos computacionales. El programa también debe graficar la derivada exacta dada por $f′ = f ′ (x)$ arriba

```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos la función f(x)
def f(x, epsilon):
    return np.sin(1 / (x + epsilon))

# Derivada exacta de f(x)
def f_prime_exact(x, epsilon):
    return -np.cos(1 / (x + epsilon)) / (x + epsilon)**2

# Aproximación de diferencias finitas para la derivada
def f_prime_fd(x, epsilon, h=1e-5):
    return (f(x + h, epsilon) - f(x, epsilon)) / h

# Parámetros
n = 100  # Número de nodos
epsilon = 1/5  # Parámetro epsilon
x_values = np.linspace(0.01, 1, n+1)  # n+1 nodos, evitamos x=0 para evitar singularidad

# Calculamos las derivadas
y_prime_fd = f_prime_fd(x_values, epsilon)  # Derivada por diferencias finitas
y_prime_exact = f_prime_exact(x_values, epsilon)  # Derivada exacta

# Graficamos ambas derivadas
plt.plot(x_values, y_prime_fd, label='Diferencias Finitas', linestyle='--')
plt.plot(x_values, y_prime_exact, label='Derivada Exacta', linestyle='-')
plt.xlabel('x')
plt.ylabel('f\'(x)')
plt.title(f'Derivada de f(x) con epsilon={epsilon}')
plt.legend()
plt.grid(True)
plt.show()
```

¿Qué tan grande tienes que elegir $n$ para que la diferencia entre estas dos funciones sea menor que $0.1$? Sugerencia: Cada función da una matriz. Cree un bucle while y use la función max de las matrices para recuperar el valor máximo y compararlos.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos la función f(x)
def f(x, epsilon):
    return np.sin(1 / (x + epsilon))

# Derivada exacta de f(x)
def f_prime_exact(x, epsilon):
    return -np.cos(1 / (x + epsilon)) / (x + epsilon)**2

# Aproximación de diferencias finitas para la derivada
def f_prime_fd(x, epsilon, h=1e-5):
    return (f(x + h, epsilon) - f(x, epsilon)) / h

# Parámetros iniciales
epsilon = 1/5  # Parámetro epsilon
tolerance = 0.1  # Diferencia máxima permitida
n = 10  # Valor inicial de n

# Inicializamos la diferencia máxima
max_difference = tolerance + 1  # Inicializamos con un valor mayor a la tolerancia

# Bucle para encontrar el valor de n adecuado
while max_difference > tolerance:
    # Generamos los valores de x
    x_values = np.linspace(0.01, 1, n+1)
    
    # Calculamos las derivadas
    y_prime_fd = f_prime_fd(x_values, epsilon)  # Derivada por diferencias finitas
    y_prime_exact = f_prime_exact(x_values, epsilon)  # Derivada exacta
    
    # Calculamos la diferencia máxima entre las dos derivadas
    max_difference = np.max(np.abs(y_prime_fd - y_prime_exact))
    
    # Imprimimos el valor actual de n y la diferencia máxima
    print(f'n = {n}, diferencia máxima = {max_difference}')
    
    # Si la diferencia es mayor a la tolerancia, incrementamos n
    if max_difference > tolerance:
        n += 1

# Mostramos el valor de n encontrado
print(f'El valor mínimo de n para que la diferencia máxima sea menor que {tolerance} es: {n}')

# Graficamos ambas derivadas para el n encontrado
plt.plot(x_values, y_prime_fd, label='Diferencias Finitas', linestyle='--')
plt.plot(x_values, y_prime_exact, label='Derivada Exacta', linestyle='-')
plt.xlabel('x')
plt.ylabel('f\'(x)')
plt.title(f'Derivada de f(x) con epsilon={epsilon} y n={n}')
plt.legend()
plt.grid(True)
plt.show()
```
