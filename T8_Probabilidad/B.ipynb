B. Considere una clase de 100 alumnos. Asigne, de manera aleatoria, sus calificaciones entre 5 y 10. Suponga que su resultado sea típico.

 a) calcule el promedio y la varianza del experimento
 
 b) repita el ejercicio 10 veces (10 cursos) 

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir el número de cursos y alumnos
num_cursos = 10
num_alumnos = 100

# Inicializar listas para guardar promedios y varianzas
promedios = []
varianzas = []

# Realizar el experimento para 10 cursos
for i in range(num_cursos):
    # Generar calificaciones aleatorias entre 5 y 10 para 100 alumnos
    calificaciones = np.random.uniform(5, 10, num_alumnos)
    
    # Calcular el promedio y la varianza
    promedio = np.mean(calificaciones)
    varianza = np.var(calificaciones)
    
    # Almacenar los resultados
    promedios.append(promedio)
    varianzas.append(varianza)

# Crear un gráfico de barras para los promedios y las varianzas
fig, ax1 = plt.subplots()

# Posiciones para las barras
indices = np.arange(num_cursos)

# Ancho de las barras
bar_width = 0.35

# Graficar barras de promedio
barras_promedio = ax1.bar(indices, promedios, bar_width, label='Promedio', color='tab:blue')

# Crear un segundo eje para la varianza
ax2 = ax1.twinx()

# Graficar barras de varianza, desplazadas en el gráfico
barras_varianza = ax2.bar(indices + bar_width, varianzas, bar_width, label='Varianza', color='tab:red')

# Configuraciones adicionales
ax1.set_xlabel('Cursos')
ax1.set_ylabel('Promedio', color='tab:blue')
ax2.set_ylabel('Varianza', color='tab:red')
ax1.set_xticks(indices + bar_width / 2)
ax1.set_xticklabels([f'Curso {i+1}' for i in range(num_cursos)])

# Título y leyendas
plt.title('Promedio y Varianza de Calificaciones en 10 Cursos')
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

print("Promedio:",promedio)
print("Varianza",varianza)

# Mostrar gráfico
plt.show()

```
