A: Calcular la probabilidad de que al tomar un conjunto de 3 alumnos de un grupo de 40, los 3 sean mujeres, si sólo hay 9 mujeres en el grupo.

La probabilidad de calcula, de la siguiente manera 
$$ \frac{\bigl\{\begin{smallmatrix} 9 \\ 3\end{smallmatrix}\bigr\}} {\bigl\{\begin{smallmatrix}40\\ 3\end{smallmatrix}\bigr\}}$$


Donde $ \bigl\{\begin{smallmatrix} n \\ k \end{smallmatrix}\bigr\} $ representa el número de combinaciones de $n$ elementos tomados $k$ a la vez.

```python
import matplotlib.pyplot as plt
import math

def combinaciones(n, k):
    """Calcula el número de combinaciones de n elementos tomados k a la vez."""
    return math.comb(n, k)

# Parámetros del problema
total_alumnos = 40
mujeres = 9
alumnos_seleccionados = 3

# Calcular combinaciones
combinaciones_mujeres = combinaciones(mujeres, alumnos_seleccionados)
combinaciones_totales = combinaciones(total_alumnos, alumnos_seleccionados)

# Calcular la probabilidad
probabilidad = combinaciones_mujeres / combinaciones_totales

# Mostrar el resultado
print(f"La probabilidad de seleccionar 3 mujeres es: {probabilidad:.4f}")

# Graficar los resultados
labels = ['Seleccionar 3 mujeres', 'Otros casos']
sizes = [probabilidad, 1 - probabilidad]
colors = ['lightcoral', 'lightskyblue']
explode = (0.1, 0)  # resaltar la primera sección

plt.figure(figsize=(8, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.2f%%', shadow=True, startangle=140)
plt.title('Probabilidad de seleccionar 3 mujeres de un grupo de 40')
plt.show()
```
